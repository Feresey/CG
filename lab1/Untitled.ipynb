```python
import numpy as np
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.models.widgets import TextInput
from bokeh.layouts import gridplot, widgetbox

from math import pi

output_notebook()
```

```python
fig = figure(tools="crosshair,pan,reset,save,wheel_zoom", title="Lab1",
             x_axis_label='x',
             y_axis_label='y')


a = TextInput(title="a", value='1.0')
b = TextInput(title="b", value='1.0')
A = TextInput(title="A", value='1.0')
B = TextInput(title="B", value='1.0')

phi = np.arange(0.*pi/180, 360.*pi/180, 0.01)


def x(a, b, phi): return a*phi - b*np.sin(phi)


def y(a, b, phi): return a - b*np.cos(phi)


source = ColumnDataSource(data=dict(x=phi - np.sin(phi),
                                    y=-np.cos(phi)))

fig.line(x="x", y="y",
         source=source,
         line_width=2,
         legend="x = a*phi - b*sin(phi)\ny = a - b*cos(phi)")

fig.xaxis.axis_label = "x"
fig.yaxis.axis_label = "y"
fig.title.text = "Lab1"


def update_source(attr, old, new):
    try:
        x = float(a.value) * np.cos(phi)
        y = float(b.value) * np.sin(phi)
        source.data = dict(x=x, y=y)
    except ValueError:
        pass

def update_dots(attr, old, new):
    global phi
    phi = np.arange(0.*pi/180, 360.*pi/180, 0.01)



for w in [a, b]:
    w.on_change('value', update_source)

for w in [A, B]:
    w.on_change('value', update_dots)

grid = gridplot([[fig, None],[a, b], [A, B]])

show(grid)
```
